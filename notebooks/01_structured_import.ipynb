# BIM RAG — Stage 1 Structured Import

Imports IFC entities and relationships into PostgreSQL using `ifc_to_db()`.
No vectors are generated in this stage.

```python
import json

from bim_rag.config import IFC_SOURCE_PATH
from bim_rag.pipeline_structured import ifc_to_db
```

## Run the structured import

Calling `ifc_to_db()` is idempotent — safe to re-run.

```python
report = ifc_to_db(str(IFC_SOURCE_PATH))
```

## Import report

```python
print(json.dumps(report, indent=2, default=str))
```

## Summary statistics

```python
print(f"Source model id : {report['source_model_id']}")
print(f"IFC schema      : {report['ifc_schema']}")
print(f"Entities        : {report['eligible_entity_count']}")
print(f"  new           : {report['entities_imported_new']}")
print(f"  updated       : {report['entities_updated']}")
print(f"Relationships   : {report['relationship_count_ifc']}")
print(f"  new           : {report['relationships_imported_new']}")
print(f"  updated       : {report['relationships_updated']}")
print(f"Members total   : {report['members_total']}")
print(f"Members resolved: {report['members_resolved']}")
print(
    f"Failures        : {report['entity_extraction_failures']} entity, "
    f"{report['relationship_extraction_failures']} rel"
)
```

## Relationship class breakdown

```python
for cls, count in sorted(report["relationship_class_counts"].items(), key=lambda x: -x[1]):
    print(f"  {cls}: {count}")
```
